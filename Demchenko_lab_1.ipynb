–ó–∞–¥–∞—á–∞ 11
–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –∑–∞–¥–∞–Ω –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ —Å–≤–æ–∏—Ö –≤–µ—Ä—à–∏–Ω:  (ùë•1,ùë¶1),(ùë•2,ùë¶2),(ùë•3,ùë¶3). –ó–Ω–∞—á–µ–Ω–∏—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Å –ø–æ–º–æ—â—å—é –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏—è. –û–Ω–∏ –º–æ–≥—É—Ç –±—ã—Ç—å –Ω–µ—Ü–µ–ª—ã–º–∏. –ù–∞–π—Ç–∏ –ø–µ—Ä–∏–º–µ—Ç—Ä –∏ –ø–ª–æ—â–∞–¥—å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞.

````python
x1, y1 = float(input("–í–≤–µ–¥–∏—Ç–µ x1: ")), float(input("–í–≤–µ–¥–∏—Ç–µ y1: "))
x2, y2 = float(input("–í–≤–µ–¥–∏—Ç–µ x2: ")), float(input("–í–≤–µ–¥–∏—Ç–µ y2: "))
x3, y3 = float(input("–í–≤–µ–¥–∏—Ç–µ x3: ")), float(input("–í–≤–µ–¥–∏—Ç–µ y3: "))

a = ((x2 - x1)**2 + (y2 - y1)**2)**(1/2)
b = ((x3 - x2)**2 + (y3 - y2)**2)**(1/2)
c = ((x1 - x3)**2 + (y1 - y3)**2)**(1/2)

perimeter = a + b + c

p = perimeter / 2

area = (p * (p - a) * (p - b) * (p - c))**(1/2)

print("–ü–µ—Ä–∏–º–µ—Ç—Ä —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞:", perimeter)
print("–ü–ª–æ—â–∞–¥—å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞:", area)
````

````python
a = float(input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ a: "))
b = float(input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ b: "))
c = float(input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ c: "))
d = float(input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ d: "))
e = float(input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ e: "))

if d - e == 0:
    print("–û—à–∏–±–∫–∞: –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ 0 –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ")
else:
    result = (a + b * c) / (2 * (d - e)) + (b ** e) % c
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç –≤—ã—Ä–∞–∂–µ–Ω–∏—è:", result)
````

–ó–∞–¥–∞–Ω–∏–µ 7 
–ù–µ –∏—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ count, –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ –≤—ã–ø–æ–ª–Ω–∏—Ç–µ:
–µ—Å–ª–∏ —Å–∏–º–≤–æ–ª * –≤ –¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç, –≤—ã–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç ¬´–Ω–µ—Ç —Å–∏–º–≤–æ–ª–∞¬ª;
–µ—Å–ª–∏ —Å–∏–º–≤–æ–ª * –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –≤ —Å—Ç—Ä–æ–∫–µ —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Ä–∞–∑, –≤—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ –∏–Ω–¥–µ–∫—Å;
–µ—Å–ª–∏ —Å–∏–º–≤–æ–ª * –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –¥–≤–∞ –∏ –±–æ–ª–µ–µ —Ä–∞–∑, –≤—ã–≤–µ–¥–∏—Ç–µ –∏–Ω–¥–µ–∫—Å –µ–≥–æ –ø–µ—Ä–≤–æ–≥–æ, –≤—Ç–æ—Ä–æ–≥–æ –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è, —É–¥–∞–ª–∏—Ç–µ –ø–µ—Ä–≤—ã–π –∏ –ø–æ—Å–ª–µ–¥–Ω–∏–π —Å–∏–º–≤–æ–ª * –∏–∑ —Å—Ç—Ä–æ–∫–∏.

````python
strl = input("–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É: ")
index = []
for i, s in enumerate(strl):
    if s == '*':
        index.append(i)
if not index:
    print("–ù–µ—Ç —Å–∏–º–≤–æ–ª–∞")

elif len(index) == 1:
    print("–ò–Ω–¥–µ–∫—Å —Å–∏–º–≤–æ–ª–∞ '*':", index[0])

else:
    print("–ò–Ω–¥–µ–∫—Å –ø–µ—Ä–≤–æ–≥–æ —Å–∏–º–≤–æ–ª–∞ '*':", index[0])
    print("–ò–Ω–¥–µ–∫—Å –≤—Ç–æ—Ä–æ–≥–æ —Å–∏–º–≤–æ–ª–∞ '*':", index[1])
    print("–ò–Ω–¥–µ–∫—Å –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —Å–∏–º–≤–æ–ª–∞ '*':", index[-1])
    strl = strl[:index[0]] + strl[index[0]+1:index[-1]] + strl[index[-1]+1:]

print("–°—Ç—Ä–æ–∫–∞ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –ø–µ—Ä–≤–æ–≥–æ –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —Å–∏–º–≤–æ–ª–æ–≤ '*':", strl)
````

````python
str1 = input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É: ")
str2 = input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä—É—é —Å—Ç—Ä–æ–∫—É: ")

str1 = str1.strip().lower()
str2 = str2.strip().lower()

if str1 == str2:
    print("–°–æ–¥–µ—Ä–∂–∞–Ω–∏–µ —Å—Ç—Ä–æ–∫ —Å–æ–≤–ø–∞–¥–∞–µ—Ç (–±–µ–∑ —É—á–µ—Ç–∞ —Ä–µ–≥–∏—Å—Ç—Ä–∞ –∏ –ø—Ä–æ–±–µ–ª–æ–≤).")
else:
    print("–°–æ–¥–µ—Ä–∂–∞–Ω–∏–µ —Å—Ç—Ä–æ–∫ –Ω–µ —Å–æ–≤–ø–∞–¥–∞–µ—Ç (–±–µ–∑ —É—á–µ—Ç–∞ —Ä–µ–≥–∏—Å—Ç—Ä–∞ –∏ –ø—Ä–æ–±–µ–ª–æ–≤).")
````

–ó–∞–¥–∞–Ω–∏–µ 12
–ü–æ –Ω–æ–º–µ—Ä—É –≥–æ–¥–∞ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –¥–∞–Ω–Ω—ã–π –≥–æ–¥ –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º. (–≥–æ–¥ —è–≤–ª—è–µ—Ç—Å—è –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º, –µ—Å–ª–∏ –µ–≥–æ –Ω–æ–º–µ—Ä –∫—Ä–∞—Ç–µ–Ω 4, –Ω–æ –Ω–µ –∫—Ä–∞—Ç–µ–Ω 100, –∞ —Ç–∞–∫–∂–µ –µ—Å–ª–∏ –æ–Ω –∫—Ä–∞—Ç–µ–Ω 400).

````python
year = int(input("–í–≤–µ–¥–∏—Ç–µ –≥–æ–¥: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("{year} - –≤–∏—Å–æ–∫–æ—Å–Ω—ã–π –≥–æ–¥")
else:
    print("{year} - –Ω–µ –≤–∏—Å–æ–∫–æ—Å–Ω—ã–π –≥–æ–¥")
````

–ó–∞–¥–∞–Ω–∏–µ 13
–î–∞–Ω—ã —Ç–æ—á–∫–∏  ùê¥=(1,1),ùêµ=(5,5),ùê∂=(1,4),ùê∑=(7,2). –ù–∞–π—Ç–∏ —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –æ—Ç—Ä–µ–∑–∫–æ–≤  ùê¥ùêµ –∏  ùê∂ùê∑.

````python
A = (1, 1)
B = (5, 5)
C = (1, 4)
D = (7, 2)

AB = (B[1] - A[1]) / (B[0] - A[0]) if B[0] != A[0] else None
CD = (D[1] - C[1]) / (D[0] - C[0]) if D[0] != C[0] else None

if AB == CD:
    print("–û—Ç—Ä–µ–∑–∫–∏ AB –∏ CD –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã –∏ –Ω–µ –∏–º–µ—é—Ç —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è.")
else:
    x = (AB * A[0] - A[1] - CD * C[0] + C[1]) / (AB - CD)
    y = AB * (x - A[0]) + A[1]

    print(f"–¢–æ—á–∫–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –æ—Ç—Ä–µ–∑–∫–æ–≤ AB –∏ CD: ({x}, {y})")
````
