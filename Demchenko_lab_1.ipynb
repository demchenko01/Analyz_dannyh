Задача 11
Треугольник задан координатами своих вершин:  (𝑥1,𝑦1),(𝑥2,𝑦2),(𝑥3,𝑦3). Значения координат определите с помощью присваивания. Они могут быть нецелыми. Найти периметр и площадь треугольника.

````python
x1, y1 = float(input("Введите x1: ")), float(input("Введите y1: "))
x2, y2 = float(input("Введите x2: ")), float(input("Введите y2: "))
x3, y3 = float(input("Введите x3: ")), float(input("Введите y3: "))

a = ((x2 - x1)**2 + (y2 - y1)**2)**(1/2)
b = ((x3 - x2)**2 + (y3 - y2)**2)**(1/2)
c = ((x1 - x3)**2 + (y1 - y3)**2)**(1/2)

perimeter = a + b + c

p = perimeter / 2

area = (p * (p - a) * (p - b) * (p - c))**(1/2)

print("Периметр треугольника:", perimeter)
print("Площадь треугольника:", area)
````

````python
a = float(input("Введите значение a: "))
b = float(input("Введите значение b: "))
c = float(input("Введите значение c: "))
d = float(input("Введите значение d: "))
e = float(input("Введите значение e: "))

if d - e == 0:
    print("Ошибка: деление на 0 невозможно")
else:
    result = (a + b * c) / (2 * (d - e)) + (b ** e) % c
    print(f"Результат выражения:", result)
````

Для двух произвольных строк провести сравнение на совпадение содержимого без учета регистра букв и начальных и конечных пробельных символов. Например, для двух строк ' HeLLO WOrlD' и 'hello WORLD ' такое сравнение должно возвращать True (содержание строк совпадает).

````python
str1 = input("Введите первую строку: ")
str2 = input("Введите вторую строку: ")

str1 = str1.strip().lower()
str2 = str2.strip().lower()

if str1 == str2:
    print("Содержание строк совпадает (без учета регистра и пробелов).")
else:
    print("Содержание строк не совпадает (без учета регистра и пробелов).")
````

Задание 12
По номеру года определите, является ли данный год високосным. (год является високосным, если его номер кратен 4, но не кратен 100, а также если он кратен 400).

````python
year = int(input("Введите год: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("{year} - високосный год")
else:
    print("{year} - не високосный год")
````

Задание 13
Даны точки  𝐴=(1,1),𝐵=(5,5),𝐶=(1,4),𝐷=(7,2). Найти точку пересечения отрезков  𝐴𝐵 и  𝐶𝐷.

````python
A = (1, 1)
B = (5, 5)
C = (1, 4)
D = (7, 2)

AB = (B[1] - A[1]) / (B[0] - A[0]) if B[0] != A[0] else None
CD = (D[1] - C[1]) / (D[0] - C[0]) if D[0] != C[0] else None

if AB == CD:
    print("Отрезки AB и CD параллельны и не имеют точки пересечения.")
else:
    x = (AB * A[0] - A[1] - CD * C[0] + C[1]) / (AB - CD)
    y = AB * (x - A[0]) + A[1]

    print(f"Точка пересечения отрезков AB и CD: ({x}, {y})")
````
